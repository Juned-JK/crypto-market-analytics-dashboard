# For automatic getting new data, we need to create a function and store that data on csv file

```python
def api_runner():
    global df
    import requests
    URL = "https://pro-api.coinmarketcap.com/v1/cryptocurrency/listings/latest"
    headers = {
        'X-CMC_PRO_API_KEY' : 'cab1fc438d284c2a98c32ab4d821608c'
    }
    params = {
        'start' : '1',
        'limit' : '5000',
        'convert' : 'USD'
    }
    
    response = requests.get(URL , headers = headers , params = params)
    data = response.json()

    df = pd.json_normalize(data['data'])
    df['Time_stamp'] = pd.to_datetime('now')

    if not os.path.isfile(r'C:\Users\juned\Videos\data analytics\end to end projects\1\Crypto_API.csv'):
        df.to_csv(r'C:\Users\juned\Videos\data analytics\end to end projects\1\Crypto_API.csv', header = 'column_names')
    else:
         df.to_csv(r'C:\Users\juned\Videos\data analytics\end to end projects\1\Crypto_API.csv', mode = 'a', header = False)
```

```python
# now lets call the function each hour for automatic data update 
import os
from time import time
from time import sleep 

for i in range(240):
    api_runner()
    print('DATA Append completed')
    sleep(3600) #3600 seconds = 1 hour
exit()    
```
